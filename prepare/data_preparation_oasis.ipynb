```python
import os
import pandas as pd
import numpy as np
import shutil
import time
import random
from pathlib import Path
import tarfile
import tensorflow as tf
```

```python
# Load .csv data
df = pd.read_csv("../data/oasis_cross-sectional.csv")
```

```python
# Get Desktop and Download path
downloads_path = str(Path.home() / "Downloads")
desktop_path = str(Path.home() / "Desktop")
dataset_folder = desktop_path + "/Oasis_train_test_val/"
horizontal_pictures = dataset_folder + "horizontal_pictures/"
```

```python
# Collect all oasis files that need to be moved and unpacked
downloads = os.listdir(downloads_path)
oasis_files = []
for files in downloads:
    if "oasis" in files and "disc" in files and "tar" in files:
        oasis_files.append(files)
    else:
        continue

# Print amount of files found
print(f"{len(oasis_files)} files found")
```

```python
# Create new folder on desktop 
if os.path.isdir(dataset_folder) == False:
    os.mkdir(dataset_folder)
    print("Successfully createt 'Oasis' folder on desktop")
else:
    print("Folder already exists")

# Move files from download to new folder
for oasis in oasis_files:
    if oasis in os.listdir(downloads_path):
        shutil.move(downloads_path + "/" + oasis, dataset_folder + oasis)
    else:
        print(f"File '{oasis}' not found")
        continue
```

```python
# Unpack files and remove packed files
for tar in os.listdir(dataset_folder):
    if ".tar" in tar:
        file = tarfile.open(dataset_folder + tar)
        file.extractall(dataset_folder)
        file.close()
        os.remove(dataset_folder + tar)
    else:
        continue
```

```python
# Create new folder for horizontal pictures
if os.path.isdir(horizontal_pictures) == False:
    os.mkdir(horizontal_pictures)
    print("Successfully createt 'Horizontal pictures' folder")
else:
    print("Folder already exists")
```

```python
for disc in os.listdir(dataset_folder):
    if "disc" in disc:
        for subfolder in os.listdir(dataset_folder + disc):
            if "OAS1" in subfolder:
                path = dataset_folder + disc + "/" + subfolder + "/PROCESSED" + "/MPRAGE" + "/T88_111"
                for image in (os.listdir(path)):
                    if "masked_gfc_tra_90" in image:
                        status = df.query(f"ID == '{image[0:len(df.ID[0])]}'")["CDR"]
                        print(status)
```
